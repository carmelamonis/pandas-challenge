```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data_complete['school_name'].value_counts())
total_schools#.count() #15 schools

total_students = len(school_data_complete) #39170
total_students

#Groups the students by schools, then grabs the first value of the budget of each school, then sums it up
total_budget = school_data_complete.groupby('school_name')['budget'].first().sum() # 82932329558
total_budget

average_math = round((school_data_complete['math_score'].mean()), 2) #78.98537145774827
average_math

average_read = round((school_data_complete['reading_score'].mean()), 2) #81.87784018381414
average_read

#Get the length of 70 or greater in math_score column, divide that by the total of students
math_passing = round(((len(school_data_complete.loc[school_data_complete['math_score'] >= 70]) / total_students)*100), 2)
math_passing #74.98

#Get the length of 70 or greater in reading_score column, divide that by the total of students
read_passing = round(((len(school_data_complete.loc[school_data_complete['reading_score'] >= 70]) / total_students)*100), 2)
read_passing #85.81


both_passing = round((len(school_data_complete.loc[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)]) / total_students) * 100, 2)
both_passing #65.17


#Create a dataframe using the data 
summary_df = pd.DataFrame([{'Total Schoools':total_schools, 'Total Students':total_students, 'Total Budget': total_budget, \
                           'Average Math Score': average_math, 'Average Reading Score':average_read, \
                           'Percent Passing Math': math_passing, 'Percent Passing Reading': read_passing, \
                           'Percent Passing Both': both_passing}])

summary_df
```

```python
#Format the students and total budget
summary_df['Total Students'] = summary_df['Total Students'].map('{:,}'.format)
summary_df['Total Budget'] = summary_df['Total Budget'].map('${:,.2f}'.format)

summary_df
```

```python
#Create a new df that is grouped by the school name
grouped_school_df = school_data_complete.groupby(["school_name"])
grouped_school_df.head()
```

```python
#Get the School Type 
#Long version before finding first()
#Get the unique value of each school type. Add each value to a list using indexing, then assign each value in the list to the column
'''def schoolType(type):
    for index in range (0, len(type)):
        return x[index][0]
    
#school_df['School Type'] = grouped_school_df['type'].unique()[0][0]
#grouped_school_df['type'].apply(lambda x: schoolType(x['type'].unique()))

#school_df.head(30)
#school_df.tail()
#grouped_school_df['type'].unique()[0][0]


#This returns an array for each row
school_df['School Type'] = grouped_school_df['type'].unique()
school_df.head(30)

#Create a list to get the value of each type
schooltypes = []
#Add the ACTUAL string value in the list
for i in range(0, len(school_df['School Type'])):
    schooltypes.append(school_df['School Type'][i][0])
        
schooltypes

#Assign that list to the school type column
school_df['School Type'] = schooltypes
school_df['School Type']
'''

#Grabs the first value of the 'type' column of each group
#This will work since each group only has one type, so it is unique
school_df['School Type'] = grouped_school_df['type'].first()
#school_df.head()
```

```python
#Get the first value of each group
school_df['Total School Budget'] = grouped_school_df['budget'].first()
```

```python
#Creating a column for Per student by grabbing values in total school budget and dividing by total students column
school_df['Per Student Budget'] = school_df['Total School Budget'] / school_df['Student Count']
```

```python
#Getting the average of each math & reading score
school_df['Average Math Score'] = grouped_school_df['math_score'].mean()
school_df['Average Reading Score'] = grouped_school_df['reading_score'].mean()
```

```python
#Use conditional to get the number of students that have a 70 or higher in math, reading, then both
#Then calculate the percentage by dividing that by the school's Student Count
passing_math = (school_data_complete.loc[school_data_complete['math_score'] >= 70, ['school_name', 'math_score']].groupby('school_name')).count()
school_df['Percentage Passing Math'] = (passing_math['math_score'] / school_df['Student Count']) * 100

passing_read = (school_data_complete.loc[school_data_complete['reading_score'] >= 70, ['school_name', 'reading_score']].groupby('school_name')).count()
school_df['Percentage Passing Reading'] = (passing_read['reading_score'] / school_df['Student Count']) * 100

passing_both = (school_data_complete.loc[((school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)), ['school_name', 'reading_score']].groupby('school_name')).count()
school_df['Percentage Passing Both'] = (passing_both['reading_score'] / school_df['Student Count']) * 100

#school_df.head()

```

```python
#Copy school_df to another df to allow for formatting
formatted_school_df = school_df.copy()
#formatted_school_df
```

```python
formatted_school_df['Total School Budget'] = formatted_school_df['Total School Budget'].map('${:,.2f}'.format)
formatted_school_df['Per Student Budget'] = formatted_school_df['Per Student Budget'].map('${:,.2f}'.format)

#Remove index name
formatted_school_df.index.name = None
formatted_school_df
```
